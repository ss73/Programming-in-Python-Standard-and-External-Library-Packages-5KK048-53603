VIII. Indexing strings
------------------------
It is possible to index a string as follows, to retrieve the character at that position: (Note that you start counting at zero.)

````python
print(original_hello[0])
print(original_hello[1])
print(original_hello[2])
print(original_hello[3])
print(original_hello[4])
print(original_hello[5])
````

It is also possible to slice strings. 
You start with the position from the index you want to keep, and then give the first index of the part of the string that you *don't* want to include in the slice. 

You can also create a slice from the start, or towards the end, by leaving out the first or the last string index respectively. 

You can also use negative numbers

````python
alphabet_str = "abcdefghijklm"

# To get 'bcd', you start with the index for 'b', i.e. 1, and end with the index after 'd', i.e. 4 (remember start by 0)
print(alphabet_str[1:4])

# From the start
print(alphabet_str[:3])

# Towards the end
print(alphabet_str[5:])

# You can also use negative numbers
print(alphabet_str[-3:])
````

IV. Mutable and immutable data types
-------------------------------------
It might look like strings are a lot like lists. They are. But there is a very important difference: 
Lists are mutable, i.e. it is possible to change their content as follows:

```
my_list = [1, 2, 3]
my_list[0] = 4
```

Strings, on the other hand, are immutable, so you cannot change their content. You cant to the assignement with and index as is okay for a list.

So if you want to change the values a lot, it is easier to work with lists. Then, when you have finished changing the content, you can change your list to a string. Note that it is easy to change a list of strings to a string by using the string method `join()¬¥



````python
my_list = ["It's", "a", "nice", "day"]
my_list[2] = "fine"

print(my_list)
joined = "-".join(my_list)
print(joined)
````
