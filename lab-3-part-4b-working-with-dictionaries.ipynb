Lab 3 - Part 4b: Working with dictionaries
==============

The dictionary is a very useful data structure. You can use it for anything where you need to do a mapping between a **key** and an **item**. The classic example is a word dictionary, where the **key** is a word you need to look up and the **item** is the translation of the word. You create the dictionary with curly brackets, and separate the key and item with `:`. As for lists and tuples, each entry is separated by a comma.

I. Creating and using a dictionary
--------------------------------

```python
swedish_animal_words = {"hund" : "dog", "katt" : "cat", "svart" : "black", "brun" : "brown"}
print(swedish_animal_words)
```

When looking up a key in the dictionary, you use square brackes, as follows:

```python
print(swedish_animal_words["hund"])
print(swedish_animal_words["svart"])
```

III. Associate several items with a dictionary key
----------------------------------------------

Often, you need to associate several items to the same key. For instance, you might want to add several English translations to each Swedish word. Then you can use a list as a data structure to store the items, as follows:

```python
swedish_animal_words_several_translations = {"hund" : ["dog"], "katt" : ["cat"], "svart" : ["black"], "brun" : ["brown"], "kanin": ["rabbit"]}
print(swedish_animal_words_several_translations)
swedish_animal_words_several_translations["kanin"].append("bunny")
print(swedish_animal_words_several_translations)
```

IV. Empty data structures and strings
--------------------------------
Note that if a data structure or string is empty, it does **not** mean that it is **None**. It is still a string/list/dictionary, just a string/list/dictionary without any content.

```python
print("None: Is None == None: " + str(None == None))
print()
print('Emty string: Is "" == None: ' + str("" == None))
print("Empty list: Is [] == None: " + str([] == None))
print("Emtpy dictionary: Is {} == None: " + str({} == None))
```

However, if you use an empty data structure or an empty list as a boolean, it will return **False**. If it has content it will instead return **True**.

```python
print(bool(None))
print(bool(""))
print(bool([]))
print(bool({}))
print()
print(bool("Example string"))
print(bool(["Example element 1", "Example element 1"]))
print(bool({"Example key" : "Example item"}))
```

This means you can use the data structures and strings in an if-clause

```python
my_dictionary = {}
if not my_dictionary:
    print("Dictionary is empty")
    
my_dictionary["A"] = 1
if my_dictionary:
    print("Items: ", my_dictionary.items())

```

V. Checking if something is in a dictionary
--------------------------------------------
The Python keyword `in` can be used for checking if something is a key in a dictionary.

```python
swedish_cities = {"Stockholm": 1617407 , "Göteborg": 607882, "Malmö": 325069 }
city_to_check = "Sundsvall"
print(f"The number of people living in {city_to_check}:")
if city_to_check in swedish_cities:
    print(swedish_cities[city_to_check])
else:
    print("is not recorded in the dictionary")
```

9) **What happens if you try `swedish_cities["Sundsvall"]` without first checking if the key is in the dictionary? What will be the result if you rely on the the .get() method instead?**

```python
city_to_check = "Malmö"
print(f"The number of people living in {city_to_check}:")
if city_to_check in swedish_cities:
    print(swedish_cities[city_to_check])
else:
    print("is not recorded in the dictionary")

try:
    city_to_check = "Sundsvall"
    print(swedish_cities[city_to_check])
except KeyError as ke:
    print("Exception:", ke, "is not in dictionary")

if swedish_cities.get(city_to_check) is not None:
    print(swedish_cities[city_to_check])
else:
    print(city_to_check, 'is not in dictionary [call get() first]')
```
